<font size="5"> <em>Pcf</em></font>  &emsp;&emsp;&emsp;  [Index](#main_index)

<font size="4"> <code>srank.Pcf(a)</code> </font>     

To define a Pcf object we use the method **Pcf** for which we need to specify 
2xn array_like of real numbers (it is assume that n>0 and that the 0-th row, refered to as the domain, consists of strictly increasing real numbers. The second row consists of values and any real value is allowed.)    

```python
# Defining ndarrays, note the use of the transpose:
A = np.array([[0, -1], [1, 3], [1.5, 4], [2,0]], dtype="double").T
B = np.array([[0, 0], [1.2, 2], [2, 4]], dtype="double").T
C = np.array([[0,1],[1,-1],[2,0.1],[2,4]], dtype="double").T

# Defining piecewise constant functions
f = sr.Pcf(A)
g = sr.Pcf(B)
h = sr.Pcf(C)

# The obtained pcfs have the following atributes:

print("content of f:")
print(f.content)

print('*' * 60)

print("content of g:")
print(g.content)

print('*' * 60)

print("content of h:")
print(h.content)
```

<a id='pcf_evaluate_plot'></a> 
<hr style="height: 0.3px;"/>

<font size="3"> <em>plot</em> and <em>evaluate</em></font></font> &emsp;&emsp;&emsp; Pcf methods    &emsp;&emsp;&emsp;  [Index](#main_index)

<font size="4"> <code>srank.plot(self, interval=None, ax=None, **kwargs)</code>  </font>     


```python
# Printing the content of both f and g
print("content of f:")
print(f.content)

print('*' * 60)

print("content of g:")
print(g.content)

print('*' * 60)


# We can evaluate a Pcf at a real number which results in a real number. 
print("f(1.1)=",f(1.1))

print('*' * 60)

# We can also evaluate it at an ndarray which results in an ndarray.
print("g([-0.5, 1.2])=", g([-0.5,1.2]))

print('*' * 60)

# Plotting the pcfs:
print("f is represented by the black graph")
print("g is represented by the red graph")
f.plot(color = "black", linewidth = 2)
g.plot(color = "red", linewidth = 1)
```

There are 6 methods related to arithmetic_operations on pcf <code>add</code>, <code>neg</code>, <code>mul</code>, <code>sub</code>, <code>pow</code>, <code>abs</code>.
&emsp;&emsp;&emsp;  [Index](#main_index)

```python
# Here is an illustration of addition
print("f is represented by the black graph")
print("g is represented by the red graph")
print("f+g+1 is represented by the yellow graph")
print("f*g is represented by the blue graph")
print("w obgtained by using np.sum([f,g,1]) is represented by the green graph")
f.plot(color = "black", linewidth = 2)
g.plot(color = "red", linewidth = 1.5)
(f + g + 1).plot(color = "yellow")
(f*g).plot(color = "blue")

# You could also apply functions such as np.sum
w = np.sum([f,g,1])
w.plot(color="green",linewidth = 0.3)
```

```python
# We can negate a pcf:
print("f is represented by the black graph")
print("-f is represented by the red graph")
f.plot(color = "black", linewidth = 2)
(-f).plot(color = "red")
```

<font size="4"> <code>srank.Pcf.integrate(self, intervals=None)</code>  </font>  

This methd can be used to integrate a pcf.

 [Index](#main_index)

```python
print("integral of f over the entier real line: ",f.integrate())

print('*' * 60)

print("integral of f over the interval [0,1.5]: ",f.integrate([0,1.5]))

print('*' * 60)

print("integral of f over the interval [0,inf]: ",f.integrate([0,np.inf]))
```

<font size="4"> <code>srank.Pcnif(a)</code> </font>     

To define a Pcnif object we use the method **Pcnif** for which we need to specify 
an array with two rows. The 0-th row describes the domain and should start with 0 and consists of strictly increasing non-negative real numbers. The 1-st row describes the values of the function and should consists of non-increasing non-negative real numbers. These conditions are assumed, however not checked! 
Pcnif object is a subobject of Pcf class.


```python
D1 = np.array([[0, 9], [0.5,7],[1,6],[2,4],[4,3],[10,1]], dtype="double").T
D2 = np.array([[0, 7], [1.5,6],[3,5],[4,4],[8,1]], dtype="double").T
D3 = [[0],[0]]
s1 = sr.Pcnif(D1)
s2 = sr.Pcnif(D2)
s3 = sr.Pcnif(D3)

print("Type of s1 + s2 is: ", type(s1 + s2))

print('*' * 60)

print("Type of s1 * s2 is: ", type(s1 * s2))

print('*' * 60)

print("Type of s1 - s2 is: ", type(s1 - s2))

print('*' * 60)

print("pcnif s1 is represented by the black graph")
print("pcnif s2 is represented by the red graph")
print("pcnif s1 + s2 is represented by the yellow graph")
print("pcnif s1 - 1 is represented by the green graph")
s1.plot(color = "black", linewidth = 2)
s2.plot(color = "red")
(s1 + s2).plot(color = "yellow")
(s1-1).plot(color = "green")
print((s1).content)
print((s1-1).content)
print((s1).integrate())
print((s1-1).integrate())
```
